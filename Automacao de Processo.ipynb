### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador



```python
#incluir nome da loja na base de vendas, no caso, ID loja tem nas duas bases
vendas = vendas.merge(lojas, on='ID Loja')
display(vendas)
```

```python
#Criar um arquivo para cada loja
for loja in lojas['Loja']:
  print(loja)
```

```python
#vou criar um dicionario(tabela) de cada loja para toda vez que eu quiser puxar sua tabela
dicionario_lojas = {}

for loja in lojas['Loja']:
  dicionario_lojas[loja] = vendas.loc[vendas['Loja']==loja,:]

  #testando lojas
display(dicionario_lojas['Rio Mar Recife'])
display(dicionario_lojas['Shopping Vila Velha'])
```

```python
#codigo pronto para toda vez que entrar um dia novo, ele conseguir calcular o indicador desse dia novo
#então vou olhar a coluna de data e pegar a maior data, pq a maior data é a mais recente 

#dia do indicador: ultima data da coluna data
dia_indicador = vendas['Data'].max()
print(dia_indicador)
print(dia_indicador.day)
print(dia_indicador.month)
print(dia_indicador.year)

#ou ainda
print('{}/{}/{}'.format(dia_indicador.day,dia_indicador.month,dia_indicador.year))
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
#identificar se a pasta ja existe. Para quando surgir loja nova e já rodar o codigo automaticamente, ja adicionar a pasta automatic 
import pathlib

caminho_backup = pathlib.Path(r'C:/Users/quel_/OneDrive/Documentos/Projeto AutomacaoIndicadores/Backup Arquivos Lojas')

#iter dir pega todos os arquivos e pastas que estao dentro do diretorio
arquivos_pasta_backup = caminho_backup.iterdir()

#vai criar um lista com o nome de todas as pastas
# lista_nomes_backup = []
# for arquivo in arquivos_pasta_backup:
#   lista_nomes_backup.append(arquivo.name)

#usando list compression
lista_nomes_backup = [arquivo.name for arquivo in arquivos_pasta_backup]
print(lista_nomes_backup)


#mas caso a loja não exista, criarei a pasta
for loja in dicionario_lojas:
  if loja not in lista_nomes_backup:
    nova_pasta = caminho_backup / loja
    nova_pasta.mkdir() #cria de fato um novo caminho/pasta

  #nome do arquivo: mes_dia_ano
  nome_arquivo = '{}_{}_{}.xlsx'.format(dia_indicador.month, dia_indicador.day,loja)
  print(nome_arquivo)
  #local_arquivo = "C:/Users.../mes_dia_nomeLoja.xlsx"
  local_arquivo = caminho_backup / loja / nome_arquivo
  
dicionario_lojas[loja].to_excel(local_arquivo)
```
